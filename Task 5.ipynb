# Visualisation

```python
# Load the dataset
df_Months_data = pd.read_csv("Months_data.csv")

# Reshape the data using pd.melt() to convert month columns into rows
df_melted = pd.melt(df_Months_data, id_vars=['States and Union Territories', 'Year'], 
                    value_vars=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
                    var_name='Month', value_name='Accidents')

# Set the style of seaborn
sb.set(style="whitegrid")

# Create a heatmap for visualizing accidents across months and states
# Pivot the data for heatmap
df_pivot = df_melted.pivot_table(index='States and Union Territories', columns='Month', values='Accidents', aggfunc='sum')

# Create a larger figure for better visibility
mpl.figure(figsize=(14, 8))

# Create the heatmap
sb.heatmap(df_pivot, cmap='Blues', annot=True, fmt="d", linewidths=.5)

# Add labels and title
mpl.xlabel('Months')
mpl.ylabel('States And Union Territories')
mpl.title('Accidents Across States/UTs By Months')

# Save the plot as an image
mpl.savefig('Accidents_States_Months_Heatmap.png', dpi=300)

# Display the plot
mpl.show()
```

```python
# Load the dataset
df_Months_data = pd.read_csv("Months_data.csv")

# Group by year and sum the 'Total Accidents' column to get yearly totals
df_yearly_totals = df_Months_data.groupby('Year')['Total  Accidents '].sum().reset_index()

# Set the style of seaborn
sb.set(style="darkgrid")

# Create the line plot
mpl.figure(figsize=(10, 6))
sb.lineplot(x='Year', y='Total  Accidents ', data=df_yearly_totals, marker="o", linewidth=2.5)

# Add labels and title
mpl.xlabel('Year')
mpl.ylabel('Total Accidents')
mpl.title('Total Accidents Over the Years (2001-2014)')

# Save the plot as an image
mpl.savefig('Total_Accidents_Over_Years.png', dpi=300)

# Display the plot
mpl.show()
```

```python
# Load the dataset
df_time_data = pd.read_csv("time_data.csv")

# Set the style of seaborn
sb.set(style="whitegrid")

# Calculate total night and day accidents across all states and union territories
total_night_accidents = df_time_data['Total Night Accidents time'].sum()
total_day_accidents = df_time_data['Total Day Accidents Time'].sum()

# Prepare the data for plotting
accident_data = pd.DataFrame({
    'Accident Time': ['Night Accidents', 'Day Accidents'],
    'Total Accidents': [total_night_accidents, total_day_accidents]
})

# Create the bar chart
mpl.figure(figsize=(8, 6))
sb.barplot(x='Accident Time', y='Total Accidents', data=accident_data, palette='Blues')

# Add labels and title
mpl.xlabel('Accident Time')
mpl.ylabel('Total Accidents')
mpl.title('Total Night vs Day Accidents')

# Save the plot as an image
mpl.savefig('Night_vs_Day_Accidents.png', dpi=300)

# Display the plot
mpl.show()

```

```python
# Load the dataset
df_time_data = pd.read_csv("time_data.csv")

# Set the style of seaborn
sb.set(style="whitegrid")

# Melt the DataFrame to have a long format suitable for seaborn
df_melted = df_time_data.melt(id_vars=['States and Union Territories', 'Year'],
                               value_vars=['Total Night Accidents time', 'Total Day Accidents Time'],
                               var_name='Accident Type',
                               value_name='Total Accidents')

# Create a larger figure for better visibility
mpl.figure(figsize=(14, 8))

# Create the bar chart
sb.barplot(x='States and Union Territories', y='Total Accidents', hue='Accident Type', data=df_melted, palette='Reds')

# Add labels and title
mpl.xlabel('States and Union Territories')
mpl.ylabel('Total Accidents')
mpl.title('Comparison of Night and Day Accidents by State')

# Rotate x-axis labels for better readability
mpl.xticks(rotation=45)

# Save the plot as an image
mpl.savefig('Comparison_of_Night_and_Day_Accidents_by_State.png', dpi=300)

# Display the plot
mpl.show()
```
